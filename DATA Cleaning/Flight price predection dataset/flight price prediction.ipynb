# Flight price precection(EDA + Feature Engineering)

```python
# Importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline   
```

```python
train_df=pd.read_excel(r"D:\Data-Science\DATA Cleaning\Flight price predection dataset\Data_Train.xlsx")
test_df=pd.read_excel(r"D:\Data-Science\DATA Cleaning\Flight price predection dataset\Test_set.xlsx")
final_df=pd.concat([train_df,test_df],axis=0,ignore_index=True)
final_df.tail()
```

# Feature Engineering

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time'].info()
```

# From Arrival Time We are going to extract Time using lambda function

```python
final_df["Arrival_Time"]=final_df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hours']=final_df['Arrival_Time'].str.split(":").str[0]
final_df['Arrival_Min']=final_df['Arrival_Time'].str.split(":").str[1]
final_df['Arrival_Hours']=final_df['Arrival_Hours'].astype(int)
final_df['Arrival_Min']=final_df['Arrival_Min'].astype(int)
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df['Dep_Hour']=final_df['Dep_Time'].str.split(":").str[0]
final_df['Dep_Min']=final_df['Dep_Time'].str.split(":").str[1]
final_df['Dep_Hour']=final_df['Dep_Hour'].astype(int)
final_df['Dep_Min']=final_df['Dep_Min'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'nan':1,'4 stops':4})
```

```python
final_df.head()
```

# we are going to drop Route Column

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h')
final_df['Duration_hour']=final_df['Duration'].str.extract(r'(\d+)').astype(int)
```

```python
final_df['Duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m')
```

```python
final_df.drop('Duration_min',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
