# 3sum

```python
def fun(l):
    res=[]
    l.sort()
    for i in range(len(l)-2):
        if i>0 and l[i]==l[i-1]:
            continue
        left=i+1
        right=len(l)-1
        while left<right:
            total=l[i]+l[left]+l[right]
            if total<0:
                left=left+1
            elif total>0:
                right=right-1
            else:
                res.append([l[i],l[left],l[right]])
                while left<right and l[left]==l[left+1]:
                    left=left+1
                while left<right and l[right]==l[right-1]:
                    right=right-1
                left=left+1
                right=right-1
    return res     
l=[-1,0,1,2,-1,-4]
fun(l)
```

# longest substring

```python
st="abcabcbb"
s=set()
for i in st:
    s.add(i)
len(s)
```
