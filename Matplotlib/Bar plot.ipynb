```python
import matplotlib.pyplot as plt
%matplotlib inline
x=["Rohit","Santosh","Ranvir","Prabhakar"]
y=[80,60,20,50]
c=['y','r','purple','b']
plt.xlabel('Names',fontsize=15)
plt.title("Student Marks")
plt.ylabel('Marks',fontstyle='italic',fontweight='bold')

plt.bar(x,y,color=c,alpha=0.5,width=.3,edgecolor='red',linewidth=3,linestyle='dotted')
```
