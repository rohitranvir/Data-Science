```python
import matplotlib.pyplot as plt
%matplotlib inline
day=[2,3,4,5,6,7]
no=[1,2,10,4,5,10]
color=[]
plt.scatter(day,no,color='r')
plt.title("Scatter plot")
plt.xlabel('days',fontsize=15,fontstyle='italic')
plt.ylabel('Number of days',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
day=[2,3,4,5,6,7]
no=[1,2,10,4,5,50]
c=['r','b','g','purple','orange','pink']
sizes=[100,200,400,600,800,1000]
plt.scatter(day,no,color=c,s=sizes,marker='*',edgecolor='r',linewidth=2)
plt.title("Scatter plot")
plt.xlabel('days',fontsize=15,fontstyle='italic')
plt.ylabel('Number of days',fontsize=15)
plt.show()
```
