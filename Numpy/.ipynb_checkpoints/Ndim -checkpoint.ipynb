## ndim in NumPy — what it is, why it’s used, and its limits
# What ndim means

ndim is a property of a NumPy array that tells you how many dimensions (axes) the array has.

It answers only one question:
“How many levels of nesting does this array have?”

# Syntax
array.ndim


No parentheses — it’s a property, not a function
# Examples with output

# 1D Array

```python
import numpy as np
a=np.array([1,2,3])
print(a.ndim)
```

# 2D Array

```python
import numpy as np
a=np.array([[1,2,3],[4,5,6]])
print(a.ndim)
```

# 3D Array

```python
import numpy as np
a=np.array([
            [
                [1,2],
                [5,6]
            ],
            [
                [1,2],
                [5,6]
            ]
             ])
print(a.ndim)
```
