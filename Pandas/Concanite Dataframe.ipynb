# CONCATENATE PANDAS DATAFRAME

How to concatenate the Pandas DataFrames in Python using the concat()
method. This will concatenate the content of the DataFrames.

Let us see an example and create two DataFrames and concatenate them:

```python
import pandas as pd
data1={
    "Name":["Rohit","Santosh","Ranvir"],
    "Age":[25,22,30],
    "Salary":[30000,25000,35000]
}
data2={
    "Name":["Rohan","Vandana","Shital"],
    "Age":[25,22,30],
    "Salary":[30000,25000,35000]
    }
df1=pd.DataFrame(data1 ,index=["Student1","Student2","Student3"])
df2=pd.DataFrame(data2, index=["Student4","Student5","Student6"])
df3=pd.concat([df1,df2])
print(df3)
```
