## 7. PANDAS DATAFRAME - HEAD()

The pandas.DataFrame.head() is used to return the first n rows.

Let us now see an example to implement the head() method in Python Pandas:

```python
import pandas as pd
data={
    "Students":["Rohit","Santosh","RAnvir"],
    "Rank":[1,2,3],
    "Marks":[95,70,80]
}
df=pd.DataFrame(data,index=["RowA","RowB","RowC"])
print(df.head())
print("\n")
print(df.head(1))
print("\n")
print(df.head(2))
```
