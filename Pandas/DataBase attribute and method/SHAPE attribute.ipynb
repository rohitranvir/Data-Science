## 4. PANDAS DATAFRAME - SHAPE

The pandas.DataFrame.shape is used to return the dimensionality of the DataFrame in
the form of a tuple.

Let us now see an example to implement the shape attribute in Python Pandas:

```python
import pandas as pd
data={
    "Students":["Rohit","Santosh","RAnvir"],
    "Rank":[1,2,3],
    "Marks":[95,70,80]
}
df=pd.DataFrame(data,index=["RowA","RowB","RowC"])
print(df.shape)
```
