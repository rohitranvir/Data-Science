## 3. PANDAS DATAFRAME - SIZE

The pandas.DataFrame.size is used to return the number of elements in the DataFrame.

Let us now see an example to implement the size attribute in Python Pandas:

```python
import pandas as pd
data={
    "Students":["Rohit","Santosh","RAnvir"],
    "Rank":[1,2,3],
    "Marks":[95,70,80]
}
df=pd.DataFrame(data,index=["RowA","RowB","RowC"])
print(df)
print(df.size)
```
