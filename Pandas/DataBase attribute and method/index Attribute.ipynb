## 5. PANDAS DATAFRAME - INDEX

The pandas.DataFrame.index is used to return the index of the DataFrame.

Let us now see an example to implement the index attribute in Python Pandas:


```python
import pandas as pd
data={
    "Students":["Rohit","Santosh","RAnvir"],
    "Rank":[1,2,3],
    "Marks":[95,70,80]
}
df=pd.DataFrame(data,index=["RowA","RowB","RowC"])
print(df.index)
```
