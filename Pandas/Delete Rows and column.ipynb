# PANDAS - DELETE ROWS/ COLUMNS

The drop() method is used in Python to delete rows/ columns in a Pandas DataFrame. It is
used to remove a particular row or column. Under the parameters of the drop( method,
mention the column you want to delete with the axis.

We will now see two examples:

1. Drop columns
2. Drop rows

# I. PANDAS - DROP COLUMNS USING DROP()

The columns are dropped using the column names. The axis is set to 1 since we want to drop
a column. The columns axis can also be used for the drop() method to remove the specified
column:

axis='columns'
or
axis = 1

Let us see an example to drop columns using the drop() method:

```python
import pandas as pd
data={
    "RollNo":[101,102,103,104],
    "Name":["Rohit","Santosh","Ranvir","Vandana"],
    "Marks":[85,90,56,98]
}
df=pd.DataFrame(data)
df["Status"]=["Pass","Fail","Fail","Pass"]
print(df)
```
