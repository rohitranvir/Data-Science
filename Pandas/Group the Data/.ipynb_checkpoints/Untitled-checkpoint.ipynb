```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["HR", "IT", "HR", "IT", "Finance"],
    "Employee": ["A", "B", "C", "D", "A"],
    "Gender":["F","F","M","M","F"],
    "Salary": [30000, 50000, 35000, 55000, 40000]
})

print(df)
```

```python
res=df.groupby("Department")["Salary"].mean()
print(res)
```

```python
res=df.groupby("Department")["Gender"].count()
print(res)
```

```python
res=df.groupby("Employee")["Gender"].count()
print(res)
```
