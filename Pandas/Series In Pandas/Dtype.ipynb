# I. PANDAS SERIES - DTYPE ATTRIBUTE

The pandas.series.dtype is used to return the datatype of the Series.

Let us now see an example to implement the dtype attribute in Python Pandas:

```python
import pandas as pd
data =[10,20,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD"])
print(df.dtype)
print(df.ndim,"D")
print(df.size)
print(df.name)
```
