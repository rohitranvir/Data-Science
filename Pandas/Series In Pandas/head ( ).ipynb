# 7. PANDAS SERIES - HEAD() METHOD

The pandas.series.head() method is used to return the first n rows of the Pandas
Series.

Let us now see an example to implement the head() method in Python Pandas:

```python
import pandas as pd
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.head(2))
```
