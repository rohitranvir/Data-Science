# 9. PANDAS SERIES - INFO() METHOD

The pandas.series.info() method is used to display the summary of the Pandas
Series.

Let us now see an example to implement the info() method in Python Pandas:

```python
import pandas as pd
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.info())
```
