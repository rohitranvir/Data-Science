# SERIES - ATTRIBUTES AND METHODS

The Series in Pandas is a one-dimensional array that uses the Series() method to create a
Series, but it also uses different built-in attributes and methods for basic functionalities.
In this lesson, let us see such attributes and methods in Python Pandas with examples:

1. dtype:  Return the dtype.
2. ndim:  Return the Number of dimensions
3. size:  Return the number of elements.
4. name:  Return the name of the Series.
5. hasnans:  Returns True if NaNs are in the series.
6. index:  The index of the series
7. head():  Return the first n rows.
8. tailQ:  Return the last n rows.
9. info0:  Display the Summary of the series



Let us understand them one by one with examples:



```python
import pandas as pd
data =[10,20,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD"])
print(df.dtype)
print(df.ndim,"D")
print(df.size)
print(df.name)
```

```python
import pandas as pd
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.ndim,"D")
```

```python
import pandas as pd
data =[10,20,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD"])
print(df.size)
```

```python
import pandas as pd
data =[10,20,30,40]
df=pd.Series(data,name="My name series",index=["RowA","RowB","RowC","RowD"])
print(df.name)
```

```python
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"]) 
print(df.hasnans)
```

```python
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.index)
```

```python
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.head(2))
```

```python
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.tail(1))
```

```python
data =[10,20,None,30,40]
df=pd.Series(data,index=["RowA","RowB","RowC","RowD","RowE"])
print(df.info())
```
