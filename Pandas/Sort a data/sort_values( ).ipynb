# I. PANDAS - SORT THE DATAFRAME

To sort the dataframe, use the sort_values() method. The default is ascending. Therefore,
you don't need to mention any value in the parameter for ascending sort.

We will sort by value i.e. the by parameter will be set to the column name by which we need
to sort the DataFrame in ascending order.

Let us now see an example:

```python
import pandas as pd
df=pd.read_csv(r"D:\Data-Science\Pandas\csv\Students.csv")
# print(df)
df_sorted=df.sort_values(by="Marks")
print(df_sorted)
```

```python
df=pd.read_csv(r"D:\Data-Science\Pandas\csv\Students.csv")
df_sorted=df.sort_values(by=["Students","Rank"] ,ascending=False)
print(df_sorted)
```
