```python
def fun(l):
    fmax=max(l)
    smax=None    
    for i in l:
        if i!=fmax:
            if smax==None:
                smax=i
            else:
                smax=max(smax,i)
    print(smax)
l=[3, 11, 4, 20, 10, 8, 1, 3, 13, 2]
fun(l)
```

```python
def change(l,index,value):
    if len(l)==index:
        return
    l[index]=value
    change(l,index+1,value+1)
    l[index]=l[index]-2
l=[0,0,0,0,0]
change(l,0,1)
print(l)
```

```python
def tower(n,src,temp,dest):
    if n==1:
        print(f"move disc {n} from {src} to {dest}")
    else:
        tower(n-1,src,dest,temp)
        print(f"move disc {n} from {src} to {dest}")
        tower(n-1,temp,src,dest)
n=3
tower(n,"src","temp","dest")
```

```python
def checking(l,index,value):
    if len(l)==index:
        return 
    l[index]=value
    checking(l,index+1,value+1)
l=[0,0,0,0,0]
checking(l,0,1)
print(l)
```

```python
def subset(s,ans,index):
    if index==len(s):
        if len(ans)==0:
            print("Null")
        else:
            print(ans)
        return
    subset(s,ans+s[index],index+1)
    subset(s,ans,index+1)
subset("abc","",0)
```

```python
def permutation(s,ans):
    if len(s)==0:
        print(ans)
        return
    for i in range(0,len(s)):
        permutation(s[:i]+s[i+1:],ans+s[i])
permutation("abc","") 
```

```python
def greedyway(i,j,m,n):
    if i==m-1 and j == n-1:
        return 1
    if i == m or j == n:
        return 0
    val1=greedyway(i+1,j,m,n)
    val2=greedyway(i,j+1,m,n)
    return val1+val2
m=3
n=3
print(greedyway(0,0,m,n))
```
