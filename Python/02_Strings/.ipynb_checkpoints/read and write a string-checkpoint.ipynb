# read string and print characters present at +ve and -ve indexes

```python
s= input("enter a number")
for i in range(len(s)):
    print(s[i])
```

```python
s=input("Enter a string")
index=0
index1=-1
while index<len(s) and index1>-(len(s)+1):
    print(s[index],s[index1])
    index+=1
    index1-=1
```
