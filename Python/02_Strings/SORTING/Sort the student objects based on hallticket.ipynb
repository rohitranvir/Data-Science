```python
class Student:
    def __init__(self, name, htno, marks):
        self.name = name
        self.htno = htno
        self.marks = marks

    def __str__(self):
        return f"({self.name}, {self.htno}, {self.marks})"

def fun(s):
    return s.htno

s1 = Student("Prakash", 333, 90)
s2 = Student("Prasanth", 666, 50)
s3 = Student("Prabas", 222, 80)
s4 = Student("Prakesh", 444, 70)

L = [s1, s2, s3, s4]

print("before sorting")
for i in L:
    print(i)

L.sort(key=fun)

print("after sorting")
for i in L:
    print(i)
```
