```python
def fun(s):
    l=s.split()
    upr=""
    for i in range(len(l)):
        upr+=(l[i][0].upper())+l[i][1:]+" "
    print(upr)
s="rohit vandana santosh prabhakar ranvir"
fun(s)
```

```python
def fun(s):
    l=[]
    for i in s.split():
        l.append(i[0].upper()+i[1:].lower())
    return " ".join(l)
s="rohit santosh ranvir vandana"
fun(s)
```
