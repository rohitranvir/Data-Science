```python
l1=[10,20,30,40,60]
l2=[100,200,300]
out=[]
n1=len(l1)
n2=len(l2)
if n1>n2:
    for i in range(n1):
        if i<n2:
            out.append(l1[i]+l2[i])
        else:
            out.append(l1[i])
print(out)
```

```python
l=[1,2,[3,4,5],[6,7,8,9]]
n1=len(l)
out=[]
for i in l:
    if type(i)==int:
        out.append(i)
    elif type(i)==list:
        out.extend(i)
print(out)        
```
