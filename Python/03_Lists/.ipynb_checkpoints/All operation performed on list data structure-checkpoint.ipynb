## Accessiong list oibject

```python
l=[10,20,30,40,50]

print(l[1])

print(l[0:len(l)])

for i in l:
    print(i)

i=0
while i!=len(l):
    print(l[i])
    i+=1
```

## Nested List objest     ==>  a list object within another list object is called nested list


```python
l=[10,20,30,40,[111,222,333],40]
print(l[4][1])
for i in l:
    print(i)
```

## List alising   =>  assiging a new referance variable for an existing list object is called as list alising

```python
l1=[10,20,30,40,50,60]
l2=l1
print(l1)
print(l2)

print(l1 is l2)


l1[2]=4000
print(l1)
print(l2)
```

## List Cloning   =>   It is used to create a duplicate copy of existing list object. it is possible by using " COPY()" method if we made any changes on origional list then it wont be reflect on duplicate list


```python
l1=[10,20,30,40,50]
l2=l1.copy()
print(l1)
print(l2)

print(l1 is l2)

l1[2]=400

print(l1)
print(l2)
```
