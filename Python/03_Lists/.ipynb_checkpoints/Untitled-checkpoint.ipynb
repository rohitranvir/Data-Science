## Generate a Even and Odd list range of 0 to 10

```python
def fun(l1,l2):
    l=[]
    for i in l1:
        if i in l2:
            l.append(i)
    return l
l1=[1,2,3,4,5]
l2=[4,5,6,7,8]
fun(l1,l2)
```

```python
def fun(l):
    n=len(l)
    for i in range(0,n):
        count=1
        for j in range(i+1,n):
            if l[i]==l[j]:
                count+=1
        if count>n/2:
            return l[i]
        
    return -1
# l=[8,3,4,8,8]
l=[8,7,6,8,6,6,6,6]
# l=[8,3,4,8,8]

fun(l)
# fun(l)
# fun(l)
```

```python
even=[]
odd=[]
for i in range(0,11):
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)
```

## Using List Comprehension

```python
even=[x for x in range(0,10+1) if x%2==0 ]
odd=[x for x in range(0,10+1) if x%2!=0]
print(even)
print(odd)
```
