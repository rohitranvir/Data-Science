```python
def fun(l):
    l1=[]
    for i in l:
        if i not in l1:
            l1.append(i)
    return l1
l=[1,2,3,4, 4,5,3]
fun(l)
```

```python
def fun(l):
    l1=[]
    i=0
    while i<len(l):
        if l[i] not in l1:
            l1.append(l[i])
        i+=1
    return l1
l=[1,2,3,4, 4,5,3]
fun(l)
```

```python
def fun(l):
    s=set()
    for i in l:
        s.add(i)
    return s
l=[1,2,3,4, 4,5,3]
fun(l)
```
