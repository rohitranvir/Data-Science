```python
def fun(l,x):
    count=0
    for i in l:
        if i>x:
            count+=1
    return count
l=[6, 2, 3, 4, 5, 1]
x=3
fun(l,x)
```

```python
def fun(l,x):
    return len([i for i in l if i>x ])
l=[6, 2, 3, 4, 5, 1]
x=2
fun(l,x)
```
