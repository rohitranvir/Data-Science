# Dictionary to list

```python
dic={'rohit': 20, 'marur': 30, 'kulkerni': 40}
keys,values=[],[]
for i in dic:
    keys.append(i)
    values.append(dic[i])
keys,values
```

2. list to string
3. list to dict
4. dict to list

```python
str1='happy happy new year year happy'
d={}
s=str1.split()
for i in range(len(s)):
    if s[i] in d:
        d[s[i]]=d[s[i]]+1
    else:
        d[s[i]]=1
print(d)
```

```python
str1='happy happy new year year happy'
s=str1.split()
d={words:s.count(words) for words in s }
print(d)
```
