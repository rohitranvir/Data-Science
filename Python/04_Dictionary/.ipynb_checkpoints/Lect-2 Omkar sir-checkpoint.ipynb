1. keys are very important
2. all inbuild function mainly working based on keys
3. index operation will not work
4. we can extract vslues by using keys only

```python
Names=['Ramesh','suresh','Satish']
age=[20,22,24]
dic={}
for i in range(len(Names)):
    dic[Names[i]]=age[i]
print(dic)
```

```python
Names=['Ramesh','suresh','Satish']
age=[20,22,24]
dic={}
for i,j in zip(Names,age):
    dic[i]=j
print(dic)
```
