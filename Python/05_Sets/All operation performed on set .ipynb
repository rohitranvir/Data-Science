1. it is group of abject of different data type
2. represented using  " { } "
3. Mutable
4. It is not indexed based
5. slicing not allowed
6. duplicated are not allowed

## Creation

## Accessing 
1. Directly
2. For each loop

```python
s={1,2,3,4,5}
print(s)           #{1, 2, 3, 4, 5}
for i in s:
    print(i)       #1
                   #2
                   #3
                   #4
                   #5
```

# Common Function on set


```python
s={1,54,56,68,23,5,64,56}
print(s)
print(max(s))   #68
print(min(s))   #1
print(sum(s))   #271


s1=sorted(s)  #[1, 5, 23, 54, 56, 64, 68]

s1[0]=1000 

s=tuple(s1)   # Type casting list into tuple
print(s)     #(1000, 5, 23, 54, 56, 64, 68)
```

# List specific Function
1. remove()   => if the item is not present then it will throw an Keyerror
2. discord()  => If the item is present then it will remove otherwise no error 
3. add()

```python
l={1,2,3,4}
l.remove(1)
l.discard(500)
l.add(50)
print(l)
```

```python
a={1,2,3,4,5}
b={45,1,5,45,65,8,}

print(a.union(b))    #{1, 2, 3, 4, 5, 65, 8, 45}

print(a.intersection(b)) #{1, 5}

print(a.difference(b))   # {2, 3, 4}

```
