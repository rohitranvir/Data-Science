# Abstraction is an OOP principle where you expose only what the object does, not how it does it.

No hype. Here’s the precise meaning, why it exists, how to implement it, and its limits.

# What abstraction actually means

1. You define a contract (what methods must exist)

2. You hide implementation details

3. Users interact with behavior, not internals

# In short:

What to do is visible, how it is done is hidden.

# Why abstraction is needed (real reason)

Without abstraction:

1. Code becomes tightly coupled

2. Changes break users

3. Large systems become unmaintainable

Abstraction reduces:

1. Dependency

2. Cognitive load

3. Change impact

# How abstraction is implemented in Python

Python uses Abstract Base Classes (ABC).

```python
from abc import ABC,abstractmethod
class bank(ABC):
    @abstractmethod
    def security(self):
        pass

class child(bank):
    def loging(self):
        print("Logging.........")
ch=child()
ch.loging()
```

# Why we do NOT write security() logic in the parent (abstract) class
Because the parent class is only a rule book, not a real object.

# Think of it like this (real-life)

Parent (abstract class) → Government rule

“Every bank must have security.”

Child (concrete class) → Actual bank

SBI, HDFC, ICICI — each has different security systems.

If the government itself builds the security system, then all banks would be forced to use the same one, which is unrealistic.

```python
from abc import ABC,abstractmethod
class parent(ABC):
    @abstractmethod
    def security(self):
        pass
class child(parent):
    def name(self):
        self.name="Rohit"
        print(self.name)
    def security(self):
        print("Now we have to write all information about security inside this METHOD")
        print("Note => if this method we did not write in child class then child class cant get accesss of parent class because this is parent class is abstract class and security is abstract method")
s=child()
s.name()
```
