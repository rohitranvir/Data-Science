```python
class Student:
    def __init__(self,name,marks):
        self.marks=marks
        self.name=name
    def averag(self,name1):
        self.name=name1
        print(self.name)
        sum=0
        for i in self.marks:            
            sum+=i
        print(f"Hii {self.name} your average is ",sum/len(self.marks))
s=Student("rohit",[95,98,65])

s.name="iron man"
name="rohit"
s.averag(name)
```
