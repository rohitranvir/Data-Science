# Encapsulation is data hiding plus controlled access it use 
# three type of objects or instances 
1. public      => We can access any where
2. protected   =>we can access outside class but we have to avoid of uses it 
3. private     =>we can not access out side class we can access only inside a class and its own methods


```python
class Student:
    def __init__(self,name):
        self.__name=name
    def show(self):
        print(self.__name)
s=Student("Rohit")
print(s.show())
```

```python
class Student:
    def __init__(self):
        self._name="Rohit"
    def show(self):
        print(self._name)
s=Student()
s.show()
print(s._name)
```
