```python
    class parent:
        def year(self):
            self.year=1990
            print(self.year)
    class child:
        def company(self):
            self.name="MERCEDIES"
            print(self.name)
            self.year()
    class Grandep(child,parent):
        def model(self):
            self.model="GLS-650"
            print(self.model)
            self.company()
    s=Grandep()
    s.model()
    
        
```
