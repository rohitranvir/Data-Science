# What polymorphism actually is

Polymorphism allows different objects to respond differently to the same method call.

The method name stays the same.
The behavior changes based on the object type.

```python
class Dog:
    def sound(self):
        print("Bark")
class Cat(Dog):
    def sound(self):
        print("mewmao")
        super().sound()
class rat(Cat,Dog):
    def sound(self):
        print("Chuchu")
        super().sound()
        
s=rat()
s.sound()

```
