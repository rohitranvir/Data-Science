```python
def bfs(graph,start):
    queue=[]
    queue.append(start)
    dfs_visited=set()
    while queue:
        if queue[0] not in dfs_visited:
            bfs_visited.add(queue[0])
        element=queue.pop(0)
        for i in graph[element]:
            if i not in bfs_visited:
                queue.append(i)
    return bfs_visited
graph={'A':['B'],'B':['A','C'],'C':['B'],'D':['E'],'E':['D','F'],'F':['E'],}
bfs(graph,'A')
```
