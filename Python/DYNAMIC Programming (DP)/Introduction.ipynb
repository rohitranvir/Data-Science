## Dynamic Programming (DP) — clear, exact definition

Dynamic Programming is a problem-solving technique used to optimize recursive or repeated computations by saving the results of subproblems so they are not solved again.

In simple words:
DP trades memory for speed.

## The core idea (nothing extra)

Dynamic Programming works only when BOTH conditions are true:

## 1️⃣ Overlapping subproblems

The same subproblem appears again and again.

Example:
To compute fib(5), you compute fib(3) multiple times.

# 2️⃣ Optimal substructure

The solution of a big problem can be built from optimal solutions of smaller problems.

If these two properties don’t exist → DP does not apply.
# two tyoe of DP
1. Memoization
2. Tabulation


```python
#USing Memoization

l=[None,None,None,None,None,None,None]
def fibo(n):
    if l[n]!=None:
        return l[n]
    if n==0 or n==1:
        return n
    else:
        l[n]=fibo(n-1)+fibo(n-2)
        return l[n]
fibo(6)
```

```python
# Using Tabulation
def fibo(n):
    l = [None] * (n + 1)
    l[0] = 0
    l[1] = 1
    for i in range(2, n + 1):
        l[i] = l[i-1] + l[i-2]
    return l[n]

print(fibo(6))
```
