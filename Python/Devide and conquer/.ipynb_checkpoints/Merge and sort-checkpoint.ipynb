# In which we take list and we merge that two list into single then we sort that list 

```python
def fun(a,b):
    c=a+b
    c.sort()
    return c
a=[10,15,20]
b=[5,6,6,3]
fun(a,b)
```

# Merge sub arrays

```python
def fun(a,low,mid,high):
    l=a[low:mid+1]
    r=a[mid+1:high+1]
    i=j=0
    print(l)
    print(r)  
    k=0
    while i<len(l) and j<len(r):
        if l[i]<r[j]:
            a[k]=l[i]
            k+=1
            i+=1
        else:
            a[k]=r[j]
            k+=1
            j+=1
    while i<len(l):
        a[k]=l[i]
        k+=1
        i+=1
    while j<len(r):
        a[k]=r[j]
        k+=1
        j+=1
L=[10,15,20,11,13]
print(L)
fun(L,0,len(L)//2,len(L)-1)
print(L)
```

# This is example for already two sorted list for merging this is not merge sort

```python
def fun(a,b):
    c=[]
    m=len(a)
    n=len(b)
    i=0
    j=0
    while i<m and j <n:
        if a[i]<b[j]:
            c.append(a[i])
            i+=1
        else:
            c.append(b[j])
            j+=1
    while i<m:
        c.append(a[i])
        i+=1
    while j<n:
        c.append(b[j])
        j+=1
    print(c)
a=[10,15,20]
b=[5,6,6,30]
fun(a,b)
```
