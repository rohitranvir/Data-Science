# Implement binary search algorithm to search for an element using recursion

```python
def binaryse(l,key,low,high):
    if low>high:
        return -1
    mid=(low+high)//2
    if key==l[mid]:
        return mid
    elif key<l[mid]:
        return binaryse(l,key,low,mid-1)
    else:
        return binaryse(l,key,mid+1,high)
    
l=[10,20,5,0,80,95,68,10,20,10]

l.sort()
print(l)
binaryse(l,20,0,len(l)-1)
```
