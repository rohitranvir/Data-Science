# 1️⃣ Linear Search
What it is

Check each element one by one until the target is found or the list ends.

# How it works

1. Start from index 0

2. Compare every element with the target

3. Stop when found or when array ends

# Example
arr = [4, 6, 1, 9, 2]
target = 9


Comparisons: 4 → 6 → 1 → 9 (found)

# Time & space

Time: O(n) (best: O(1), worst: O(n))

Space: O(1)

# Advantages

1. Works on unsorted data

2. Very simple

3. No preprocessing

# Drawbacks (real)

1. Slow on large datasets

2. Wastes comparisons

3. Not scalable

```python
def linearse(l,key):
    for i in range(len(l)):
        if key==l[i]:
            return i
    return -1
l=[10,20,5,0,80,95,68]
linearse(l,120)
```

# Implement linear search algorithm to return first and second algorithm

```python
def lsearchfirstandsecoccur(l,key):
    Tl=[]
    for i in range(len(l)-1):
        if l[i]==key:
            Tl.append(i)
    return Tl[:2]
l=[10,20,5,0,80,95,68,10,20,10]
lsearchfirstandsecoccur(l,10)
```

```python
def lsearchlastoccur(l,key):
    Tl=[]
    for i in range(len(l)-1):
        if l[i]==key:
            Tl.append(i)
    return Tl[-1:-2:-1]
l=[10,20,5,0,80,95,68,10,20,10]
lsearchfirstandsecoccur(l,10)
```
