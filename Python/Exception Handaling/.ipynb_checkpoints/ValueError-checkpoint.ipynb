```python
try:
    x=int(input("Enter a value"))
    y=10/x
    print(y)
except ValueError:
    print("Value must be integer")
else:
    print("")
```

```python

balance=500
withdrawal=int(input("Enter a amount"))
if withdrawal>balance:
    raise ValueError("Amount must be less that balance")
balance-=withdrawal
print("Remaining amount",balance)
```
