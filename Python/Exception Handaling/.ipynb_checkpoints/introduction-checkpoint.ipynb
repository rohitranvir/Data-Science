## Exception handaling is a error control mechanism whitch is used to deals with Handaling error that occure after program run its purpose is not to hide error but shows the error and prevent program from crashing

#  Why exce ption exit

2. missing files

4. Null access

# Without Exception handaling

# Exception Handaling is help program to fail safely for silently

```python
try:
    x= int(input("Enter a number"))
    y = 10
    Z=y/x
    print(Z)
except ZeroDivisionError:
    print("Division by zero is not allowed")
except ValueError:
    print("Invalid input Integer Required")
else:
    print("Exceusion Successfull")
finally:
    print('Cleanup successfull')
```
