```python
def Eular_graph_or_not(graph):
    for i in graph:
        if len(graph[i])%2 !=0:
            return False
    return True
graph={
    "A":['B','C'],
    "B":['A','C'],
    "C":['B','D'],
    'D':['A','C']
}
if Eular_graph_or_not(graph):
    print("This graph is Eular Graph")
else:
    print("This graph is not Eular Graph")
```
