# Creation of linked list

```python

class node:
    def __init__(self,data):
        self.data=data
        self.next=None
head=node(10)
print(head.data)
print(head.next)
```

```python
head.next=node(20)
print(head.next.data)
```

```python
head.next.next=node(30)
print(head.next.next.data)
```

## printing a list

```python
def printlist(head):
    temp=head
    if temp==None:
        print("List is empty")
        return
    while temp!=None:
        print(temp.data,end=" => ")  
        # print("=>")
        temp=temp.next    
    print(None)
```

```python
printlist(head)
```
