```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
head=node(10)
head.next=node(20)
head.next.next=node(30)
print(head.data)
print(head.next.data)
print(head.next.next.data)
```

```python
def printlist(head):
    temp=head
    if temp==None:
        temp=print("List is empty")
        return
    while temp!=None:
        print(temp.data,end=" => ")
        temp=temp.next
    print(None)
```
