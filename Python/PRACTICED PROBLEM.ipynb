```python
def fun(l):
    fmax=max(l)
    smax=None    
    for i in l:
        if i!=fmax:
            if smax==None:
                smax=i
            else:
                smax=max(smax,i)
    print(smax)
l=[3, 11, 4, 20, 10, 8, 1, 3, 13, 2]
fun(l)
```

```python
def change(l,index,value):
    if len(l)==index:
        return
    l[index]=value
    change(l,index+1,value+1)
    l[index]=l[index]-2
l=[0,0,0,0,0]
change(l,0,1)
print(l)
```

```python
def tower(n,src,temp,dest):
    if n==1:
        print(f"move disc {n} from {src} to {dest}")
    else:
        tower(n-1,src,dest,temp)
        print(f"move disc {n} from {src} to {dest}")
        tower(n-1,temp,src,dest)
n=3
tower(n,"src","temp","dest")
```
