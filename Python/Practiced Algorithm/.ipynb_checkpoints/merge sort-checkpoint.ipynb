```python
def merg(a,b):
    c=[]
    i,j=0,0
    while i<len(a) and j<len(b):
        if a[i]<b[j]:
            c.append(a[i])
            i+=1
        else:
            c.append(b[j])
            j+=1
    c.extend(a[i:])
    c.extend(b[j:])
    print(c)

a=[1,3,5,7,9]
b=[2,4,6,8,10]
merg(a,b)
```
