```python
def tower(n,src,temp,dest):
    if n==1:
        print(f"Move disc 1 from {src} to {dest}")
    else:
        tower(n-1,src,dest,temp)
        print(f"Move disc {n} from {src} to {dest}")
        tower(n-1,temp,src,dest)
n=int(input("Enter numbe r of disc"))
tower(n,"src","temp","dest")
```
