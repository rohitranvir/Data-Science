## N-Queens Problem
What it is

The N-Queens problem asks you to place N queens on an N√óN chessboard so that no two queens attack each other.
A queen attacks along rows, columns, and diagonals.
The task is to find one valid arrangement or all possible arrangements.

Why the problem exists (real constraint)

If you place queens randomly or row-by-row without checks, conflicts appear immediately:

1. Same column ‚Üí direct attack

2. Same diagonal ‚Üí indirect attack

This problem exists to demonstrate constraint satisfaction, search pruning, and exponential complexity management.

Core constraints

For any queen at position (row, col):

1. No other queen in the same column

2. No other queen on the same main diagonal (row - col)

3. No other queen on the same anti-diagonal (row + col)

Standard solution approach: Backtracking

Backtracking builds the solution row by row, abandoning partial boards as soon as a conflict is detected.

Why backtracking works

1. You never explore invalid branches fully

2. Conflicts are detected early

3. Time complexity drops drastically compared to brute force

Reality check

1. Still exponential in the worst case
