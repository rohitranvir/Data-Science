```python
import re
text="my age is 21"
result=re.match(r"my",text)
print(result.group())
```

```python
import re

text = "My age is 21 and my brother is 18"

matches = re.finditer(r"\d+", text)
for m in matches:
    print(m.group(), m.start(), m.end())
```

```python
import re
string ="price:123$"
pt="[0-9.]+"
print(re.findall(pt,string))
```
