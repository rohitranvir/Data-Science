```python
import random
def selection_sort(l):
    for i in range(len(l)-1):
        min_i=i
        for j in range(i+1,len(l)):
            if l[min_i]<l[j]:
                min_i=j
        l[i],l[min_i]=l[min_i],l[i]
    return l
l=[]
for i in range(10):
    l.append(random.randint(100,999))
print(l)
selection_sort(l)
```
