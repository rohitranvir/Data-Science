```python
class stack:
    # constructure for initialization
    def __init__(self):
        self.stk=[]
    # isempty Operation
    def isempty(self):
        return len(self.stk)==0
    # Size of Stack
    def size(self):
        return (len(self.stk))
    
s=stack()
print(s.isempty())
print(s.size())      
```
