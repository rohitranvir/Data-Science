# Stack Data Structure in Python
### What is a Stack?

A stack is a linear data structure that follows LIFO — Last In, First Out.
The last element added is the first one removed.

Think of it like a pile of plates:

1. You add a plate on top → push

2. You remove the top plate → pop

3. You can only see or access the top plate → peek

4. You cannot access elements in the middle directly.

# Core Operations
1. Operation	=> Meaning
2. push(x)	    => Add element to top
3. pop()	    => Remove top element
4. peek()	    => View top element
5. is_empty()	=> Check if stack is empty
6. size()	    => Number of elements

All of these are O(1) operations in a proper stack implementation.
