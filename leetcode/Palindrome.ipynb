```python
import re
def isPalindrome(s):
    # s1=s.replace(" ","")
    s2=re.findall(r"[a-zA-Z0-9]",s)
    print(s2)
    s2="".join(s2).lower()
    start =0
    end=len(s2)-1
    while start<=end:
        if s2[start]!= s2[end]:
            return False
        else:
            start+=1
            end-=1
    return True                  
    
s="race a car"
isPalindrome(s)
```
