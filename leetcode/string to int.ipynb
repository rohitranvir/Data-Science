```python
import re
def myAtoi(s):
    
    pt="[^a-zA-Z]"    
    
    r=""    
    for i in s:
        if s[0].isalpha():
            return 0
        else:
            r+=i
    s1=int("".join(re.findall(pt,r)))
    s2=abs(s1)
    sign =-1 if s1<0 else 1
    return s2*sign
myAtoi("words and 987")
```
